## __while__
Example : while machine still running, take data from machine log files to import database

__Example__: _Are you __HUMAN__ ?_

````python
list_of_acceptance=['Y','N','y','n']

# first ask if you are human? 
your_answer=input('Are you Human? Y/N: ')

# if input NOT in [Y,N,y,n], repeat  
while your_answer not in list_of_acceptance:
    
    your_answer=input('Ask again, are you Human? Y/N: ')

# after get right answer
print('Thanks for the correct answer')
````

````python
cnt = 1

while cnt > 0:
    print(cnt)
    cnt = cnt + 1
    
    # be carefult with endless loops!
# please change the condition to 'cnt >0'
# while cnt > 0
````

````python
cnt = 1

while True:
    print(cnt)
    cnt = cnt + 1
    if cnt > 10:
        break
    
    # be careful with endless loops!
````

## __FOR__ Loops

for ```variable``` in ```sequence```: <br>
&emsp; body_of_loop that has set of statements <br>
&emsp; which requires repeated execution

Here ```<variable>``` is a variable that is used for iterating over a ```<sequence>```. On every iteration it takes the next value from ```<sequence>```  until the end of sequence is reached.

````python
c_list = [8,7,6,5,4,3,2,1]
````

````python
for i in c_list:
    print(i)
````

````python
for i in c_list:
    sq = i * i
    print(sq)
print("Program Finished!")
````

````python
for i in range(8):
    print(c_list[i])
````
